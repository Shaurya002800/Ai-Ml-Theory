*A Naive Bayes classifier*

A Naive Bayes classifier is a simple but fast machine learning algorithm used for classification tasks. It is based on Bayes' Theorem, a mathematical formula for calculating the probability of an event happening based on prior knowledge. 

Why is it called "Naive"?

It is considered "Naive" because it makes a massive, often unrealistic assumption: it believes every feature in your data is completely independent of the others. 

Example: If you are identifying a fruit as an "Apple," the model looks at features like "Red," "Round," and "4-inch diameter".
The Naive Part: It assumes being "Red" has nothing to do with being "Round," even though in reality, these traits often go together. It simply treats each piece of evidence as its own separate "vote".

Top Use Cases

Spam Filtering: One of its most famous uses—sorting "Spam" from "Ham".

Sentiment Analysis: Deciding if a review is positive or negative.

Real-time Prediction: Because it is so fast and requires very little training data, it is great for quick, "on-the-fly" decisions.

*What is the primary difference between L1 (Lasso) and L2 (Ridge) regularization*

1. L1 Regularization (Lasso)

->What it does: It can shrink the weights of less important features all the way to zero.

->The Result: It effectively removes those features from the model.

->Best for: Feature Selection. If you have 100 features but suspect only 10 actually matter, Lasso will "kill off" the useless 90. 

2. L2 Regularization (Ridge)

->What it does: It shrinks the weights of all features, but never to zero. It just makes them very, very small.

->The Result: Every feature stays in the model, but their "influence" is weakened so no single feature can dominate the prediction.

->Best for: Dealing with Multicollinearity (when your features are highly correlated). It keeps all the info but prevents the model from overreacting to any one part. 


The "Penalty" Math (Simple Version)

L1 (Lasso): Adds a penalty based on the Absolute Value of the weights(w)
This creates a "sharp" corner in the math that forces weights to hit zero.

L2 (Ridge): Adds a penalty based on the Square of the weights (w^2)
This creates a "curved" penalty that gets smaller and smaller but never quite touches zero. 

*Data Leakage*

Data leakage is essentially "cheating" in machine learning. It happens when your model accidentally gets a "sneak peek" at information during training that it wouldn't actually have in the real world when making a prediction.

Two Main Types of Leakage

1. Target Leakage (The "Future Data" Trap)

This happens when you include a feature that is only available after the event you are trying to predict has already occurred. 

->Example: You want to predict if a patient has pneumonia.

->The Leak: You include a feature called "Took Antibiotics".

->Why it's a leak: In the real world, a doctor gives antibiotics after diagnosing pneumonia. If your model sees the patient took antibiotics, it "knows" they have the illness, but this doesn't help a doctor who needs a diagnosis before prescribing treatment.

2. Train-Test Contamination (The "Mixing" Trap)

This happens when information from your test set (the data you're supposed to keep hidden) "leaks" into your training set during preparation. 

->Example: You decide to "Scale" your data (make all numbers between 0 and 1).

->The Leak: You calculate the average (mean) of the entire dataset and then split it into Train and Test sets.

->Why it's a leak: Your training set now "knows" the average of the test set. The model has technically already "seen" the test data through that average, so its test scores will be artificially high

*Feature Scalling*

Feature scaling is a preprocessing step in machine learning that adjusts the range of your numerical features so they are all on a comparable scale

->Distance-Based Models: Algorithms like K-Nearest Neighbors (KNN) and K-Means Clustering measure the physical distance between points. If one feature is in the thousands and another is in the tens, the larger one will completely dominate the calculation.

->Gradient Descent: In models like Linear Regression or Neural Networks, scaling helps the optimization process "step" more smoothly toward the best answer, leading to faster convergence.

->PCA: Principal Component Analysis is sensitive to variance. If features aren't scaled, those with naturally larger ranges will wrongly appear more important. 

3 Main Techniques: 

| Technique                     | How it Works                                                                          | Best Use Case                                                                                                            |
| ----------------------------- | ------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------ |
| **Normalization (Min-Max)**   | Rescales data values to a range between **0 and 1**.                                  | Neural Networks or when the **data distribution is unknown**.                                                            |
| **Standardization (Z-Score)** | Centers data around a **mean of 0** with **standard deviation of 1**.                 | Most ML models like **SVM, Logistic Regression**, especially when data follows a **Gaussian (bell curve)** distribution. |
| **Robust Scaling**            | Uses **Median and IQR (Interquartile Range)** instead of mean and standard deviation. | When the dataset contains **extreme outliers** that could distort normal scaling methods.                                |


When can you skip it?

->Tree-Based Models: Algorithms like Decision Trees, Random Forest, and XGBoost are "scale-invariant." They split data based on "greater than" or "less than" rules, so the actual range doesn't matter.

->Naive Bayes: Since it treats each feature independently, it doesn't need them to be on the same scale
