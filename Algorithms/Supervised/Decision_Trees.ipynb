
What it is: Builds a tree of if-else rules by repeatedly splitting data on the feature that gives the most "information gain" (reduces impurity the most).

Where it's used: Customer segmentation, medical diagnosis, anywhere you need explainability.

Why use it: Fully interpretable ("if age > 30 AND income > 50k, then approved"), handles both numerical and categorical features, no need to scale data.

Limitations: Prone to overfitting (memorizing training data). Solved by ensemble methods below.
