What it is: Finds the hyperplane that maximally separates two classes. Uses the "kernel trick" to handle non-linear data by projecting it into higher dimensions.

Where it's used: Text classification, image classification (pre-deep learning era), bioinformatics (gene expression classification).

Why use it: Very effective in high-dimensional spaces (like text), theoretically solid, works well with small datasets.

Limitations: Slow on large datasets, hard to interpret, sensitive to feature scaling.
