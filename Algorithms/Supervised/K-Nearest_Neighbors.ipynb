What it is: To classify a new point, find its K nearest neighbors in training data and take a majority vote.

Where it's used: Recommendation systems (users similar to you liked this), anomaly detection, image recognition (simple cases).

Why use it: Zero training time, naturally handles multi-class, intuitive.

Limitations: Slow at prediction time (must compare to 
all training points), sensitive to irrelevant features, needs feature scaling.
