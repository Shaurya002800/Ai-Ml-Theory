Instead of finding centroids, DBSCAN asks: "is this point in a dense neighborhood?" It classifies every point as one of three types:

->Core point — has at least minPts neighbors within radius epsilon

->Border point — within epsilon of a core point, but not itself a core point

->Noise point — neither core nor border. These are outliers.

Clusters form by connecting core points that are within epsilon of each other.

Why this is powerful

DBSCAN can find clusters of any shape — rings, crescents, blobs, spirals. It doesn't need you to specify K. And it explicitly identifies outliers. K-Means would fail completely on a dataset shaped like two concentric rings; DBSCAN handles it perfectly.

Parameters

epsilon (ε) — the radius of neighborhood to consider,
minPts — minimum points required to form a core point

Choosing these: use a k-distance graph (plot distance to k-th nearest neighbor for all points, sorted). The "elbow" in that plot is a good epsilon.

Where DBSCAN is used

->Geospatial clustering (find hotspots on a map — restaurants, crime locations)

->Anomaly detection (noise points = anomalies, no threshold tuning needed)

->Image segmentation

->Scientific data with irregular cluster shapes (astronomy, biology)

Strengths

->No need to specify K

->Finds arbitrarily shaped clusters

->Robust to outliers — explicitly marks them

->Clusters of varying density work better than K-Means

Weaknesses

->Struggles when clusters have very different densities

->Epsilon is hard to choose in high-dimensional data (curse of dimensionality)

->Slower than K-Means — O(n²) naive, O(n log n) with spatial indexing
