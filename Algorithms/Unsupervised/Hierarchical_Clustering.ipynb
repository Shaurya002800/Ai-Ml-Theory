Starts with every single point as its own cluster. Then repeatedly merges the two closest clusters until only one remains. This builds a dendrogram — a tree showing the full merging history. You then "cut" the tree at any height to get your desired number of clusters.

Linkage methods — how to measure "closest clusters"

This is the key design choice:

->Single linkage — distance between clusters = distance between their two closest points. Creates long, chained clusters. Sensitive to outliers.
->Complete linkage — distance = between two farthest points. Creates compact, roughly equal clusters.
->Average linkage — distance = average of all pairwise distances between the two clusters. Good balance.
->Ward's method — merges clusters that minimize the increase in total WCSS. Usually the best default — tends to produce equal-sized compact clusters.

The dendrogram

The dendrogram lets you visually inspect the natural cluster structure. The y-axis is the distance at which clusters were merged. Cut the tree where there's a big "jump" in merge distance — that's your natural number of clusters.

Where it's used

->Biology (phylogenetic trees, gene expression analysis)
->Document organization
->When you want to explore cluster structure at multiple levels of granularity
->Social network analysis

Strengths

->No need to specify K upfront
->Produces a full hierarchy you can explore
->Deterministic — same result every time
->Works well for small to medium datasets

Weaknesses

O(n² log n) to O(n³) — very slow on large datasets
Greedy — once two points are merged, they can't be separated
Hard to scale beyond ~10,000 points
